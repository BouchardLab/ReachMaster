```python
def cone_trials(pos_vector,xd_vector,yd_vector,zd_vector, zeros_vector):
    images = []
    for v in pos_vector:
        fig = plt.figure(figsize=(8, 8))
        ax = fig.add_subplot(111, projection='3d')
        # Make cube surface
        #Xc, Yc = np.meshgrid(np.linspace(-0.1, 0.1, 15), np.linspace(-0.1, 0.1, 15))
        #Zc_bot =   np.reshape(np.linspace(-0.1, -0.1, 15*15), Xc.shape)
        #Zc_top =  np.reshape(np.linspace(0.1, 0.1, 15*15), Xc.shape)
        #Zc_mid = np.reshape(np.linspace(0.0, 0.0, 15*15), Xc.shape)
        #ax.plot_surface(Xc, Yc, Zc_bot, alpha=0.4, zorder=0)
        #ax.plot_surface(Xc, Yc, Zc_top, alpha=0.4, zorder=0)
        #ax.plot_surface(Xc, Yc, Zc_mid, alpha=0.4, zorder=0)
        #X1,Y1,Z1 = get_cone_reaching_vars()
        #ax.plot_wireframe(X1, -Y1, Z1)
        ax.plot(xd_vector, zeros_vector, zeros_vector, linewidth = 10, color='r', label = '1-D Xdim')
        ax.plot(zeros_vector, yd_vector, zeros_vector, linewidth = 10, color='g', label = '1-D Ydim')
        ax.plot(zeros_vector, zeros_vector, zd_vector, linewidth = 10, color='b', label = '1-D Zdim')
        ax.scatter(v[0],v[1],v[2], s = 300, c = 'gold', label= 'Handle Varying in XYZ' )
        plt.legend(fontsize="small", borderpad = 0.5, loc='upper right', labelspacing = 0.5, handletextpad=0.6 )
        filename = '3d_trialn' + str(v[0]) + '.png'
        ax.set_xlabel('M')
        ax.set_ylabel('M')
        ax.set_zlabel('M')
        ax.set_xlim(-0.1,0.1)
        ax.set_ylim(-0.1,0.1)
        ax.set_zlim(-0.1,0.1)
        plt.title('Handle Positions: 2019 3-D Experiments in a 3-D Cone')
        plt.savefig(filename)
        plt.close()
        images.append(imageio.imread(filename))
    return images



def get_cone_reaching_vars():
    radi = 0.15
    height = 0.2
    a=0.025 #x
    b=0.075 #y
    c=0 #z
    choose=max(radi,height)
    # Set up the grid in polar
    theta = np.linspace(0,2*np.pi,90)
    r = np.linspace(-.1,choose,50)
    T, R = np.meshgrid(-theta, r)
    # Then calculate X, Y, and Z
    X = R * np.cos(T) + a
    Y = R * np.sin(T) + b
    Z = (np.sqrt((X-a)**2 + (Y-b)**2)/(radi/height)) + c
    return X, Y, Z


def xform_coords_euclidean(r,theta,phi):
    x=r*np.sin(theta)*np.cos(phi)/400
    y=r*np.sin(theta)*np.sin(phi)/400
    z=r*np.cos(phi)/400
    return x,y,z


cone_positions = [[["r"],["thetay"],["thetay.1"]],
[10,0.740480489693061,0.740480489693061],
[10,0.740480489693061,-0.740480489693061],
[10,-0.740480489693061,0.740480489693061],
[10,-0.740480489693061,-0.740480489693061],
[40,0.740480489693061,0.740480489693061],
[40,0.740480489693061,-0.740480489693061],
[40,-0.740480489693061,0.740480489693061],
[40,-0.740480489693061,-0.740480489693061],
[20,0,0] ]
cone_array = np.zeros((9,3))
for z, c in enumerate(cone_positions):
    if z > 0:
        cone_array[z-1,:] = xform_coords_euclidean(c[0], c[1], c[2]) # CM units
cone_ims = cone_trials(cone_array, xpoints, ypoints, zpoints, zeros_spacing)
make_gif_from_plots(cone_ims, 'cone.mp4')
```
